Crie dois novos DataFrames, um contendo a as linhas de reclamações de pessoas do sexo masculino e outra do sexo feminino. Use estes DataFrames para responder:

Qual é a proporção de reclamações entre homens e mulhores? (total de reclamações de homens dividido pelo total de reclamações de mulheres)

```python
# Resposta:
df_homem = data[data['TIPO']=='RECLAMAÇÃO'][['TIPO','SEXO']]
df_homem = df_homem[df_homem['SEXO']=='M']

df_mulher = data[data['TIPO']=='RECLAMAÇÃO'][['TIPO','SEXO']]
df_mulher = df_mulher[df_mulher['SEXO']=='F']

homens = len(df_homem)
mulheres = len(df_mulher)

if(homens > mulheres):
    print("Homens reclamaram mais")
else:
    print("Mulheres reclamaram mais")
    
print("Proporção: ", homens/mulheres)
```

Concatene os dois DataFrames criados acima para criar um DataFrame chamado `df_todos`. Verifique se este DataFrame tem o mesmo núimero de linhas do DataFrame lido do CSV.

```python
# Resposta:
df_todos = pd.concat([df_homem,df_mulher])

if(len(df_todos) == len(df_homem)+len(df_mulher)):
    print("Os DataFrames foram unidos corretamente!")
else:
    print("Algo deu errado!")
df_todos
```
